## Ejercicio 23

```r
x <- seq(-5, 5, length = 50)
y <- seq(-5, 5, length = 50)

z <- outer(x, y, function(x,y) {
    r <- (2*x^2 +3*x*y + 4*y^2)/(3*x^2 + 5*y^2)
    return (r)
})
```

```r
persp(x, y, z,
      theta  = 15,                      # Rotacion en el eje Z (en grados)
      phi    = 40,                      # Rotacion en el eje X (en grados)
      expand =  1,                      # Escalar los valores de Z
      border = '#42caaa',               # Color del borde (En hexadecimal)
      col    = c('#004A94', '#42caaa'), # Color de valores (En hexadecimal)
      shade  =  0.5)                    # Contraste de sombra proyectada)
```

## Ejercicio 24

$f(x,y) = \frac{xy^{3}}{x^{2} + y^{6}}$

```r
x <- seq(-10, 10, length = 50)
y <- seq(-10, 10, length = 50)

z <- outer(x, y, function(x,y) {
    r <- (x*y^3)/(x^2 + y^6)
    return (r)
})
```

```r
persp(x, y, z,
      theta  = 15,                      # Rotacion en el eje Z (en grados)
      phi    = 40,                      # Rotacion en el eje X (en grados)
      expand =  1,                      # Escalar los valores de Z
      border = '#333333',               # Color del borde (En hexadecimal)
      col    = c('#bbddff', '#ddbbff'), # Color de valores (En hexadecimal)
      shade  =  0.5)                    # Contraste de sombra proyectada)
```

## Ejercicio 27

$f(x,y) = e^{\frac{1}{x-y}}$

```r
x <- seq(2, 10, length = 50)
y <- seq(1, 10, length = 50)

z <- outer(x, y, function(x,y) {
    r <- exp(1/(x-y))
    return (r)
})
```

## Ejercicio 28

$f(x,y)=\frac{1}{(1 - x^{2} - y^{2})}$

```r
x <- seq(-2, 2, length = 50)
y <- seq(-2, 2, length = 50)

z <- outer(x, y, function(x,y) {
    r <- 1/(1-x^2-y^2)
    return (r)
})
```

```r
persp(x, y, z,
      theta  = 15,                      # Rotacion en el eje Z (en grados)
      phi    = 40,                      # Rotacion en el eje X (en grados)
      expand =  1,                      # Escalar los valores de Z
      border = '#333333',               # Color del borde (En hexadecimal)
      col    = c('#bbddff', '#ddbbff'), # Color de valores (En hexadecimal)
      shade  =  0.5)                    # Contraste de sombra proyectada)
```
