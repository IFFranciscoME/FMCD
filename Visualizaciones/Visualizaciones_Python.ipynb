```python
from sympy.plotting import plot3d
from sympy import symbols

# -- Caso 1
# -- -- f(x,y) = x^2 + y - 1
# -- -- g(x,y) = x - 2y^2

x, y = symbols('x y')
funcion1 = 'x**2 + y - 1'
funcion2 = 'x - 2*y**2'
```

$f(x,y) = x^{2} + y - 1$

$g(x,y) = x - 2y^{2}$

```python
plot3d((funcion1, (x, -5, 5), (y, -5, 5)),
       (funcion2, (x, -5, 5), (y, -5, 5)))
```
